# **Analysis of Visitor Preference Dataset**

```python
df = pd.read_excel('../Data/Visitors Preference Dataset.xlsx')
df.head()
```

```python
import ast

activities = []

def extract_list_item(string):
    converted_list = ast.literal_eval(string)
    for i in range(len(converted_list)):
        activities.append(converted_list[i])
    df['size of activities'] = len(converted_list)
```

```python
df['Preferred Activities'].apply(extract_list_item)
```

```python
print(activities)
```

```python
activities_set = set(activities)
print(activities_set)
```

```python
len(activities_set)
```

```python
from collections import Counter

activity_count = Counter(activities)
print(activity_count)
```

```python
bucket_list_destinations = []

def extract_bucket_list_destinations(string):
    converted_list = ast.literal_eval(string)
    for i in range(len(converted_list)):
        bucket_list_destinations.append(converted_list[i])
    df["size of bucket list"] = len(converted_list)

df['Bucket list destinations Sri Lanka'].apply(extract_bucket_list_destinations)

print(len(bucket_list_destinations))
```

```python
unique_bucket_list_destinations = set(bucket_list_destinations)
print(unique_bucket_list_destinations)
```

```python
bucket_item_count = Counter(bucket_list_destinations)
print(bucket_item_count)
```

```python
bucket_item_count_df = pd.DataFrame.from_dict(bucket_item_count, orient='index').reset_index()
bucket_item_count_df.columns = ['Destination', 'Count']

activity_count_df = pd.DataFrame.from_dict(activity_count, orient='index').reset_index()
activity_count_df.columns = ['Activity', 'Count']
```

```python
import matplotlib.pyplot as plt

activity_count_df = activity_count_df.sort_values(by='Count', ascending=False)

activity_count_df.plot(kind='bar', x='Activity', y='Count', color='blue', figsize=(15, 6))

plt.title('Activity Count')
plt.xlabel('Activity')
plt.ylabel('Count')

plt.show()
```

```python
import matplotlib.pyplot as plt

bucket_item_count_df = bucket_item_count_df.sort_values(by='Count', ascending=False)

bucket_item_count_df.plot(kind='bar', x='Destination', y='Count', color='red', figsize=(30, 6))

plt.title('Bucket Item Count')
plt.xlabel('Bucket Item')
plt.ylabel('Count')
plt.show()
```

### **Fill Missing Data**

```python
df['rating'].fillna(df['rating'].mean(), inplace=True)
```

there are outliers in user_rating_total so we cannot use mean wiyhout cleaning outleirs

```python
def remove_outliers_iqr_multi(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df
```

```python
multi_outliers_removed_df = remove_outliers_iqr_multi(df, columns=["user_ratings_total"])

df['user_ratings_total'].fillna(multi_outliers_removed_df['user_ratings_total'].mean(), inplace=True)
df.shape
```

```python
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(15, 6))

sns.histplot(df['user_ratings_total'], bins=50, color='blue', kde=True)

plt.title('User Ratings Total Distribution', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.show()
```
